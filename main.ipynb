# Color Classification

The color classification task is to predict the distinct color category based on the three input values for RGB (R: Red, G: Green, B: Blue) color code. Each of these three numbers is basically an integer ranging from 0 to 255 and these combined Red, Green & Blue values are used to generate a distinct solid color. The dataset contains 11 basic color categories for classification. These basic colors are Red, Green, Blue, Yellow, Orange, Pink, Purple, Brown, Grey, Black and White.

Import necessary libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import accuracy_score
```

Load data into a dataframe

```python
df = pd.read_csv('./data/data.csv')
df
```

 Encode "label" column into numerical data using sklearn LabelEncoder

```python
le = LabelEncoder()
le.fit(df['label'])
df['label'] = le.transform(df['label'])
df['label']
```

Split the data into train and test datasets with 80:20 ratio

```python
features = df.copy()
labels = features.pop('label')
features = np.array(features)
labels = np.array(labels)

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
print('X_train',X_train)
print('y_train',y_train)
print('X_test',X_test)
print('y_test',y_test)

```

Install and import tensorflow library 

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.layers import Dense
```

Build and compile the sequential model. The model has a input, a output and a hidden layer. Activation function of the hidden layer is relu. The model is optimized by the Adam optimizer.

```python
model = tf.keras.Sequential([
  Dense(3),
  Dense(7,activation='relu'),
  Dense(11)
])

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model(X_train[0:2],y_train[0:2])
model.summary()
```

Train the model on the training dataset for 500 epochs

```python

model.fit(X_train, y_train, epochs=500)
```

Predict the color category for the testing data and save the result in csv file

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
probs = probability_model.predict(X_test)
predictions = np.argmax(probs, axis=1)
result = np.concatenate((X_test, np.expand_dims(predictions, axis=1)), axis=1)
result_df = pd.DataFrame(result,columns=['red', 'green', 'blue','label'])
result_df.to_csv('.prediction.csv',index = False)
result_df
```

Evaluate the accuracy score of model using the predicted data

```python
accuracy = accuracy_score(y_test, predictions)
print('Accuracy',accuracy)
```
